```python
data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "Ford"],
    "counts": [3, 7, None, 8, 1, 9, None]
}

df_missing = pd.DataFrame(data_with_missing)

print(df_missing)
```

```python
df_missing.isnull().sum()
```

```python
df_missing.dropna() # bunda qaysi qatorda cars yoki countsdan birortasini qiymati bo'lmasa o'chirib tashlaydi
```

```python
df_missing.dropna(how="all") # bunda qaysi qatorda cars yoki countsdan ikkitasini ham qiymati bo'lmasa o'chirib tashlaydi
```

```python
df_missing.dropna(axis=1) # bunda qaysi columnda birorta qiymati bo'lmasa o'sha columnni o'chirib tashlaydi
```

```python
df_missing.dropna(subset="counts") # bunda counts bo'yicha qaysi qatorda qiymati bo'lmasa o'sha qatorni o'chirib tashlaydi
```

```python
df_missing.fillna({"cars":"unknown","counts":0}) # bu datalari yo'q bo'lgan joylarni biz kiritgan ma'lumot bilan almashtirib ko'rsatadi
```
